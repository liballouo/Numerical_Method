```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
# define the compute function
def function(x):
    f = np.exp(np.sin(x))
    return f
```

```python
# exact value of function f integrated from -2 to 4
exact_value = 6.208758035711
# determine the interal of integration
# a : lower bound of integration
# b : upper bound of integration
a = 0.
b = np.pi
```

```python
def simpson_rule(num_of_grid_points):
    # define the number of intervals
    n = num_of_grid_points
    # the step h is defined by h=(b-a)/(n-1)
    h = (b-a)/(n-1)
    # generate data points
    x = np.linspace(a, b, n)
    f = function(x)
    # calculate the integration
    I_simp = (h/3)*(f[0] + 2*sum(f[2:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1])
    
    return I_simp
```

```python
# determine the interval of the grid points
# x : lower bound
# y : upper bound
x = 3
y = 13
# num_of_tests is defined by (y-x)/2
num_of_tests = (y-x)//2
# number of grid points of tests 
grid_points_of_tests = np.arange(x, y, 2)
# initialization
I_simp = np.zeros(num_of_tests)
error_simp = np.zeros(num_of_tests)

for i in range(num_of_tests):
    # calculate the integration by Simpson Rule
    I_simp[i] = simpson_rule(2*i+3)
    # error is defined by |I_simp-exact_value|
    error_simp[i] = abs(I_simp[i]-exact_value)
```

```python
print("minimum error: {0}".format(error_simp.min()))
plt.figure(figsize = (12, 8))
plt.plot(grid_points_of_tests, error_simp)
plt.xlabel("grid points")
plt.ylabel("error")
plt.show()
```
