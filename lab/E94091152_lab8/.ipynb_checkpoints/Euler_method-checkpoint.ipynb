```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
# define function f
f = lambda t, s: np.exp(-1*2*t) # the R.H.S of the ODE
# define time interval delta_t
delta_t = 0.1
# numerical grids t
t = np.arange(0, 1+delta_t, delta_t)
# initial condition s_0
s_0 = -1/2
```

```python
# Forward Euler method
# define tmp solution
s = np.zeros(len(t))
s[0] = s_0

# solve the solutions iteratively
for i in range(0, len(t)-1):
    s[i+1] = s[i]+delta_t*f(t[i], s[i])
```

```python
plt.figure(figsize = (12, 8))
plt.plot(t, s, '-', label='Forward Euler Method')
plt.plot(t, -1*0.5*np.exp(-2*t), 'r--', label='Exact')
plt.title('Euler v.s. Exact')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.rc('legend', fontsize=10)
plt.legend(loc="lower right")
```
