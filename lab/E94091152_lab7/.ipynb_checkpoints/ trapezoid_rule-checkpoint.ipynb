```python
import numpy as np
import math
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
def function(x):
    # define the function
    f = 4*(x**2) + (1/3)*x + 5
    return f
```

```python
# determine the interal of integration
# a : lower bound of integration
# b : upper bound of integration
a = -2
b = 4
# define the data points on the interval
x_curve = np.linspace(a, b, 1000)
# exact area under the function curve on the interval
exact_area = 128
```

```python
# set the color order of the area
area_colors = ['darkorange','forestgreen','royalblue']

def trapezoid_rule(num_of_grid_points):
    # n : number of grid points
    # h : integration operator delta_x
    n = num_of_grid_points
    h = (b - a) / (n - 1)
    # define data points x_i
    x = np.linspace(a, b, n)
    # define f as the values on the data points
    f = function(x)
    # integrate the function on the interval by trapezoid rule
    integration_trapezoid_area = math.floor(sum(h/2*(f[0:n-1]+f[1:n])) * 1000) /1000
    # plot the function curve and trapezoid area 
    # plot the function curve on the interval
    plt.plot(x_curve , function(x_curve), linewidth=2, c = 'r')
    # polt the trapezoid area
    plt.fill_between(x, f, where = np.where(x)[0]%3 != 2 , color=area_colors[0], alpha=0.5)
    plt.fill_between(x, f, where = np.where(x)[0]%3 != 0 , color=area_colors[1], alpha=0.5)
    plt.fill_between(x, f, where = np.where(x)[0]%3 != 1 , color=area_colors[2], alpha=0.5)

    # define the error
    error = round(exact_area-integration_trapezoid_area, 3)
    # print the result
    print("I_trap: {0}".format(integration_trapezoid_area))
    print("err_trap: {0}".format(error))
```

```python
trapezoid_rule(3)
```

```python
trapezoid_rule(11)
```

```python
trapezoid_rule(21)
```
